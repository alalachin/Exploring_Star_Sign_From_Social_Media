```python
from dcard import Dcard
import time
import json
import random
```

```python
article = []
dcard = Dcard()
# 板的名字
forum_name = 'horoscopes'

# 1. 欲抓取自訂數量NUM的文章
NUM = 3000
metas = dcard.forums(forum_name).get_metas(num=NUM)

# 2. 抓取整個星座版的文章
# metas = dcard.forums('horoscopes').get_metas(num=dcard.forums.infinite_page)

suc_num = len(metas)
print ("get all: "+ str(suc_num))

for i in range(suc_num):
    article.append(dcard.posts(metas[i]).get())
    # time.sleep(2)
    print (i)
    time.sleep(random.uniform(2,5))

    
dataset = []
count = 0
for i in range(suc_num):
    time.sleep(random.uniform(2,5))
    try:
        temp = article[i].result()[0]
        if(len(temp)>10):
            print ('im here')
            dataset.append(temp)
        else:  # 若一文章的key少於10個，代表文章已被刪除，則跳過不抓
            print ("-------"+str(count)+"is deleted.")
        count+=1
    except:
        print (str(count)+" goes to except")
        count+=1
        continue
```
